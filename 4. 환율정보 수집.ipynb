# iframe 태그 수집
- iframe = 창을 통해서 다른 서버에 존재하는 데이터를 보여줄 때 사용
- 보이는 데이터가 실제 다른 공간에 존재.
- 해결책 : 실제 데이터가 있는 주소로 찾아가야한다.
- src경로가 상대경로 인 경우에는 반드시 절대경로로 수정해 줄것!
- http:~~~ / 첫번째 슬레쉬가 있는 곳까지가 메인 서버의 url
- 데이터를 수집할 때 수집이 안되는 경우 크게 3가지
- 1번째 = 선택자가 잘못된 경우
- 2번째 = iframe태그에 감싸진 경우
- 3번째 = 수집이 불가능한 사이트인 경우
