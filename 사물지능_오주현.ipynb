# 문제4
멜론차트에서 1~100위까지 정보를 수집하는 프로그램을 작성하시오.

```python
import numpy as np
import pandas as pd

import requests as req
from bs4 import BeautifulSoup as bs

req.get('https://www.melon.com/chart/index.htm')

# F12로 network에서 document로 User-Agent 찾아서 넣어줘야함
user_agent = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"}

res = req.get('https://www.melon.com/chart/index.htm', headers = user_agent)
res.status_code

soup = bs(res.text, "lxml")
soup

title = soup.select("div.ellipsis.rank01 > span > a")

titles = [i.text for i in title]

singer = soup.select("span.checkEllipsis")
singers = [i.text for i in singer]

rankings = list(range(1, 101))

dic = {'순위': rankings, '가수':singers, '노래제목':titles}
melon = pd.DataFrame(dic)
melon.to_csv('melon_top_100.csv',encoding='utf-8')

melon
```

# 문제5
자동제어를 통해 한솥 페이지에 접속하고 전체 메뉴의 메뉴명 가격정보를 수집할 수 있는 프로그램을 작성하세요 

```python
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
import pandas as pd

driver = wb.Chrome()
driver.get("https://www.hsd.co.kr/menu/menu_list")
time.sleep(2)
try :
    for i in range(100) :
        btn = driver.find_element(By.CSS_SELECTOR,"#btn_more > span > a")
        btn.click()
        time.sleep(1)
        
except :
    print("더보기 완료")

title = driver.find_elements(By.CSS_SELECTOR,"div > div.item-text > h4")
price = driver.find_elements(By.CSS_SELECTOR,"div > div.item-text > div > strong")

title_list = []
price_list = []
for i in range(len(title)) :
    title_list.append(title[i].text)
    price_list.append(price[i].text)

dic = {"메뉴명" : title_list, "가격" : price_list}
hansot = pd.DataFrame(dic)
hansot.to_csv("한솥.csv",encoding="utf-8")

driver.close()

hansot
```
